```python
import pandas as pd
test=pd.read_csv('/content/drive/MyDrive/project bot/Testing.csv')
train=pd.read_csv('/content/drive/MyDrive/project bot/Training.csv')
```

```python
X_train=train.iloc[:,:-1]
Y_train=train.iloc[:,-1]
X_test=test.iloc[:,:-1]
Y_test=test.iloc[:,-1]
```

```python
# Encode the labels
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(Y_train)
y_test_encoded = label_encoder.transform(Y_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import SGD
```

```python
# neural network
model = Sequential()
model.add(Dense(128, input_shape=(len(X_train.columns),), activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(set(Y_train)), activation='softmax'))
```

```python
# Compile the model
sgd = SGD(learning_rate=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='sparse_categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```python
# Train the model
hist = model.fit(X_train, y_train_encoded, epochs=100, batch_size=8, validation_data=(X_test, y_test_encoded), verbose=1)

```

```python
# Save the model
model.save('medical_diagnosis_bot.h5')
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test_encoded)
print(f'Test accuracy: {accuracy}')
print(f'Loss: {loss}')
```
